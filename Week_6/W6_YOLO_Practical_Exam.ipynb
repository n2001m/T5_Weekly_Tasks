## Process Video Frames and Identify Vehicles Left the Road Form Each Street

For each frame in the video, use the YOLO model to detect and track vehicles. You will need to write a loop that processes each frame and applies the YOLO model.

In each frame, after detecting the vehicles, determine whether they exited the road via the top, bottom, left, or right street. You can use the positions of the bounding boxes provided by YOLO to do this.

- The video should display bounding boxes around the detected objects. ✅

- The video should display the number of vehicles that exited via the right street.

```python
top_C = 0
bottom_C = 0
left_C =0
right_C = 0
```

```python
from ultralytics import solutions
region_points = [(950, 130), (950, 260), (1200, 130), (1200, 260)]

top_Counter = solutions.ObjectCounter(
    view_img=True,
    reg_pts=region_points,
    names=model.names,
    draw_tracks=True,
    line_thickness=2,
)
```

```python
bottom_Counter = solutions.ObjectCounter(
    view_img=True,
    reg_pts=[(580, 760), (580, 1020), (900, 760), (900, 1020)],
    names=model.names,
    draw_tracks=True,
    line_thickness=2,
)
```

```python
left_Counter = solutions.ObjectCounter(
    view_img=True,
    reg_pts=[(200, 330), (200, 550), (400, 330), (400, 550)],
    names=model.names,
    draw_tracks=True,
    line_thickness=2,
)
```

```python
right_Counter = solutions.ObjectCounter(
    view_img=True,
    reg_pts=[(1350, 420), (1350, 660), (1500, 420), (1500, 660)],
    names=model.names,
    draw_tracks=True,
    line_thickness=2,
)
```

```python
ret = True
while ret:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.rectangle(frame, (top[0], top[1]), (top[2], top[3]), (255, 0, 0), 2)
    cv2.rectangle(frame, (bottom[0], bottom[1]), (bottom[2], bottom[3]), (255, 0, 0), 2)
    cv2.rectangle(frame, (left[0], left[1]), (left[2], left[3]), (255, 0, 0), 2)
    cv2.rectangle(frame, (right[0], right[1]), (right[2], right[3]), (255, 0, 0), 2)

    results = model.track(frame, persist=True)
    for result in results:
        for box in result.boxes:
            #BBOX, confidence, car ID, class ID
            bbox = box.xyxy[0].cpu().numpy()
            conf = box.conf[0].cpu().numpy() if box.conf is not None else 0.0
            id = int(box.id[0].cpu().numpy()) if box.id is not None else -1
            classID = int(box.cls[0].cpu().numpy()) if box.cls is not None else -1

            #BBOX with center
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
            car_x = (x1 + x2)//2
            car_y = (y1 + y2)//2
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.circle(frame, (car_x, car_y), 5, (0, 0, 255), -1)

            #Label
            label = f'conf: {conf} ID: {id} class: {classID}'
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)

    top_C = top_Counter.start_counting(frame, results)
    bottom_C = bottom_Counter.start_counting(frame, results)
    left_C = left_Counter.start_counting(frame, results)
    righr_C = right_Counter.start_counting(frame, results)

    output.write(top_C)
    output.write(bottom_C)
    output.write(left_C)
    output.write(righr_C)

    #My name
    cv2.putText(frame, 'Narjes Alkhars', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2, cv2.LINE_AA)

    output.write(frame)

cap.release()
output.release()
```
